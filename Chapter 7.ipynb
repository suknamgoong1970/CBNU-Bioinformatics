## T-test

30개 이상의 정규분포를 따르는 두 가지 서로 독립적인 데이터의 평균이 다른지를 살펴본다.   
이전에 사용한 Arabidopsis와 Rice 의 전체 Proteome의 길이의 평균이 다른지를 T-test를 통하여 검정한다. 

지난 시간에 


```python
#5,6장의 내용을 합쳐서 DNA와 Protein file을 넣으면 데이터프레임화하는 함수를 만듬
import pandas as pd
from pyfaidx import Fasta

def MolWeight(proteinsequence):
# 분자량을 구하는 함수 
# proteinsequence 에 아미노산 서열을 받고 
# MW 에는 각각의 아미노산과 분자량의 딕셔너리가 있다. 
    MW = {'G':57.05,
          'A':71.09,
          'S':87.08,
          'T':101.11,
          'C':103.15,
          'V':99.14,
          'L':113.16,
          'I':113.16,
          'M':131.19,
          'P':97.12,
          'F':147.18,
          'Y':163.18,
          'W':186.21,
          'D':115.09,
          'E':129.12,
          'N':114.11,
          'Q':128.14,
          'H':137.14,
          'K':128.17,
          'R':156.19,
          'X':0,
          'J':0}
    MolecularWeight=0
    for aa in proteinsequence:
        MolecularWeight=MolecularWeight+MW[aa]
    return(MolecularWeight)

def mid_extract (original, search1, search2):
    if (search1 in original) & (search2 in original):
        return(original.split(search1)[1].split(search2)[0])
    else:
        return('') 


#5강에서 했던 단백질 서열에서부터 정보 추출 
def DNAProtein(Proteinfile,DNAfile):
   
    AraProtein= Fasta(Proteinfile)
    proteinkey = []
    proteindescription = []
    proteinsequence = []

    for protein in AraProtein.keys():
        proteinkey.append(protein)
        proteindescription.append(AraProtein[protein].long_name.split(protein+' ')[1].split(' [')[0])
        proteinsequence.append(str(AraProtein[protein]))

    proteins = pd.DataFrame(data={'key':proteinkey,'description':proteindescription,'sequence':proteinsequence})
    AraSequence= Fasta(DNAfile)

    #5강에서 했던 DNA CDS 시퀀스 설명으로부터 정보 추출

    DNAsequence = []
    DNAproteinid = []
    DNAlocusid = []

    for Sequence in AraSequence.keys():
        name = AraSequence[Sequence].long_name
        DNAsequence.append(str(AraSequence[Sequence]))
        DNAproteinid.append(mid_extract(name,'[protein_id=',']'))
        DNAlocusid.append(mid_extract(name,'[locus_tag=',']'))

    DNASeq = pd.DataFrame(data={'protein':DNAproteinid,'sequence':DNAsequence, 'locus':DNAlocusid})

    #이 내용을 DNAprotein 이라는 데이터프레임으로 저장 
    DNAprotein = pd.merge(DNASeq,proteins, left_on='protein',right_on='key', how='left')
    DNAprotein['MW']=DNAprotein['sequence_y'].apply(MolWeight)
    return(DNAprotein)
    
#5,6장의 내용을 합쳐서 DNA와 Protein file을 넣으면 데이터프레임화하는 함수를 만듬    
Rice = DNAProtein('GCA_000004655.2_ASM465v1_protein.faa','GCA_000004655.2_ASM465v1_cds_from_genomic.fna')
Arabidopsis = DNAProtein('GCA_000001735.2_TAIR10.1_protein.faa','GCA_000001735.2_TAIR10.1_cds_from_genomic.fna')
```

## ANOVA와 Post-hoc analysis 

T-test는 두 개의 그룹의 비교를 위한 것이고, 만약 3개 이상의 집단을 비교하기 위해서는 다음과 같은 절차를 따라야 한다. 

ANOVA : 각 그룹간의 차이가 존재하는가를 검정 
Post-hoc analysis : 사후검정을 통하여 각각의 집단 간의 유의적인 변화를 조사한다. 

만약 3개 이상의 그룹간의 차이가 존재하는 경우, 각각의 그룹간의 비교를 수행하여 각각의 차이가 존재하는지를 본다. 
이때 단순한 T-test 분석을 수행하면 안되고, 다중 비교에 따라서 P-Value를 보정하는 테스트를 사용해야 함.

사후 분석 (Post-Hoc Analysis)를 수행하기 위해서는 데이터 형식을 바꾸어야 한다.  

각각의 데이터와 데이터의 타입을 만든 다음과 같은 Dataframe을 만든다. 

```python
A = pd.DataFrame(Arabidopsis['MW'])
A['type']='Arabidopsis'

R = pd.DataFrame(Rice['MW'])
R['type']='Rice'

Y = pd.DataFrame(Yeast['MW'])
Y['type']='Yeast'

C = pd.DataFrame(Corn['MW'])
C['type']='Corn'

combined = pd.concat([A,R,Y,C])
combined
```

combined 라는 이름으로 데이터프레임을 만들고, Tukey's HSD 라는 Post-hoc analysis 방법을 이용하여 사후검정을 수행한다. 


```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison

mc = MultiComparison(combined['MW'],combined['type'])
results = mc.tukeyhsd()
print (results)
```

```python
import pandas as pd

proteome = pd.read_csv('protein.tsv',sep='\t')
proteome = proteome.fillna(0)
proteome = proteome.drop(['Gene ID'],axis=1)
proteome.set_index('Gene Name',inplace=True)
tissues = proteome.transpose()
tissues
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
mean = pd.DataFrame(np.mean(tissues))
filtered = mean[mean[0]>np.quantile(mean,0.2)].index.to_list()
filtereddata = tissues[filtered]
```

```python
from sklearn.preprocessing import normalize
normal = normalize(filtereddata)
data_normalized = pd.DataFrame(normal, columns=filtereddata.columns, index = filtereddata.index)
```

```python
data_normalized
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(20, 15))  
plt.title("Dendrograms")  


from sklearn.preprocessing import StandardScaler
temp = filtereddata.to_numpy()
scaler = StandardScaler()
scaler.fit(temp)
normal = scaler.transform(temp)


dend = shc.dendrogram(shc.linkage(normal, method='centroid'),labels=filtereddata.index,leaf_font_size=10)
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
pca.fit(normal)
```

```python
principalComponents = pca.fit_transform(normal)
```

```python
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf.index = filtereddata.index
```

```python
principalDf
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
p=sns.scatterplot(x="principal component 1",y="principal component 2",data=principalDf)

for index in principalDf.index.to_list():
    x = principalDf.loc[index]['principal component 1']
    y = principalDf.loc[index]['principal component 2']
    p.text(x+0.01,y,index,size='large',color='black')
```

단백질과 조직을 동시에 Clustering 하여 조직에 특이적으로 발현되는 단백질의 목록을 찾는다.   
Seaborn의 clustermap 기능을 사용하자. 

```python
#z_score=1 이라는 옵션을 설정하면 데이터의 표준화를 수행한다. 
#데이터 표준화는 조직 내의 데이터를 평균값을 뺀 후, 표준편차로 나누어 샘플 내에서의 상대적인 값을 비교할 수 있게 한다. 

import seaborn as sns
g=sns.clustermap(filtereddata,z_score=1)
```

```python
#
# Cluster Map에서 순서가 바뀐 항목을 받아오고, 이를 이용하여 원래의 데이터를 변환한다. 
#

row = g.dendrogram_row.reordered_ind
col = g.dendrogram_col.reordered_ind

row
```

```python
temp = filtereddata.to_numpy()
#데이터를 표준화한다. 
#조직별로 분포하는 단백질 수치를 평균을 0으로하여 변환한다. 

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(temp)
normal = scaler.transform(temp)


#normal 에 있는 데이터의 순서를 Clustermap의 순서대로 재배열한다. 
u = normal[row,:][:,col]
sns.heatmap(u)
u
```

```python
#조직의 순서를 재배열한다. 

index = filtereddata.index.to_numpy()[row]
index
```

```python
#유전자의 순서를 재배열한다. 
gene = filtereddata.columns.to_numpy()[col]
gene
```

```python
clustered = pd.DataFrame(data=u, index=index,columns=gene)
clustered
```
